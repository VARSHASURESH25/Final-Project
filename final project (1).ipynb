Found out the countries represented by country codes and created a country name column for easy readability and understanding there by finding out which country has the largest number of restaurants according to the data given

```python
def convert_to_usd(row):
    country = row["Country Name"]
    amount = row["Average Cost for two"]

    if "UAE" in country:
        return amount * 0.27
    elif "India" in country:
        return amount / 83    
    elif "United States" in country:
        return amount         
    elif "United Kingdom" in country:
        return amount * 1.22  
    elif "Brazil" in country:
        return amount * 0.20  
    elif "South Africa" in country:
        return amount * 0.054 
    elif "Turkey" in country:
        return amount * 0.034 
    elif "Indonesia" in country:
        return amount * 0.000054 
    elif "Sri Lanka" in country:
        return amount * 0.0031 
    elif "Phillipines" in country:
        return amount * 0.017 
    else:
        return None
```

```python
data["Amount in USD"] = data.apply(convert_to_usd, axis=1)
```
